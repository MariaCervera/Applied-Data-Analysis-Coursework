# Taming Text

```python
%matplotlib inline
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import nltk # if some package missing run nltk.downloads()
```

### Exercise 1

```python
# Load the data and obtain text
path = 'hillary-clinton-emails/Emails.csv'
file = pd.read_csv(path) # read csv email files
raw_text = ' '.join(np.array(file.RawText)) # concatenate all text contained in RawText field
```

```python
# 1) GENERATE WORD CLOUD FROM RAW CORPUS 

# Generate a word cloud image
wordcloud = WordCloud().generate(raw_text)

# Display the generated image
plt.imshow(wordcloud)
plt.axis("off")
plt.title('Raw text')

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(raw_text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# 2) TEXT-PREPROCESSING PIPELINE

# Tokenization (identification of words)
tokens = nltk.word_tokenize(raw_text[:5000])
nltk_text = nltk.Text(tokens)

# Term normalization
words = [w.lower() for w in nltk_text]
vocab = sorted(set(words))

# Stopword removal
# Stemming 

pre_processed_text = vocab
```

```python
# 3) GENERATE WORD CLOUD FROM PRE-PROCESSED CORPUS

# Generate a word cloud image
wordcloud = WordCloud().generate(pre_processed_text)

# Display the generated image
plt.imshow(wordcloud)
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(pre_processed_text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.title('Pre-processed text')
plt.show()
```
